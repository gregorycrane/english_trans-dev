```python
import re 

def addentry(curdict,curent):
    if(curent in curdict):
        curdict[curent] = curdict[curent] + 1
    else:
        curdict[curent]  = 1
f = open('xenophon_1907-sent2.xml')
innotes = 0
refs = {}
notes = {}
curpage = 'tba'
intext = 0
for l in f:
    l = re.sub('\s+$','',l)
    if(re.search('<text',l)):
        intext = 1
    m = re.search('<pb n="([^"]+)"/>', l)
    if(m):
        curpage = m[1]
        #print(curpage)
        innotes = 0

    if(re.search('<note type="foot',l)):
        innotes = 1

    #print(innotes,l)
    m = re.search('<note type="footnote">\s*([0-9\*]+)\s*',l)
    if(m):
        curkey = curpage + '-' + m[1]
        addentry(notes,curkey)
        #print(curkey,l)
        l = re.sub('<note type="footnote">\s*([0-9\*]+)\s*','<note n='+curpage + '.\g<1>">',l)

    if( innotes or not intext):
        continue


    m = re.search('<ref n="([^"]+)"',l)
    if(m):
        addentry(refs,m[1])
    refpat = '([a-z][,\.:;”\?!]*|[\.\?,][”"]|,\)|>;)([0-9\*]+)'
    
    while(re.search(refpat,l)):
        m = re.search(refpat,l)
        if(m):
            #print(curpage,m[2],l)
            curkey = curpage + '-' + m[2]
            addentry(refs,curkey)
            #print('ref',curpage,m[2],l)
        l =  re.sub(refpat,'\g<1><ref n="'+curkey+'"/>',l,1)
    

f.close()

for foo in refs:
    if(not foo in notes):
        print('missing note',foo)
for foo in notes:
    if(not foo in refs):
        print('missing',foo)
```
