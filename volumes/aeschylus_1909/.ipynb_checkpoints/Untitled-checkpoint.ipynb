```python
import re
import os

gkadds = {}
for foo in os.listdir('gk'):
    keyname = re.sub('(.+)\.txt','\g<1>',foo)
    #print(keyname)
    gkadds[keyname] = '<quote rend="blockquote">\n'
    f = open('gk/' + foo)
    for l in f:
        l = re.sub('\s+$','',l)
        gkadds[keyname] = gkadds[keyname] + '<l>' + l + '</l>\n'

    f.close()
    gkadds[keyname] = gkadds[keyname] + '</quote>\n'

f = open('aeschylus_1909.xml')
outf = open('aeschylus_1909-2.xml','w')
for l in f:
    l = re.sub('\s+$','',l)
    print(l,file=outf)
    m = re.search('(ENG_BELL_01_fig[0-9]+)',l)
    if(m):
        print(m[1],gkadds[m[1]])
        print(gkadds[m[1]],file=outf)

f.close()
outf.close()
#print(os.listdir('gk'))
```
